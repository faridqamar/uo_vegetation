```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
testimage = ['108', '000', 'north']
```

```python
# read confusion matrix from metrics file

def read_matrix(filename, imgtype):
    f = open(filename, "r")
    f1 = f.readlines()
    if imgtype == '000':
        lines = np.arange(1,9)
    else:
        lines = np.arange(1,10)
    
    for line in lines:
        Nums = []
        s1 = f1[line].replace('[', ' ')
        s2 = s1.replace(']', ' ')
        for num in s2.split():
            try:
                Nums.append(int(num))
            except ValueError:
                pass

        if line == 1:
            CM = Nums
        else:
            CM = np.vstack((CM, Nums))
    
    return CM
```

```python
# create confusion matrix plot 

def plot_confusion_matrix(ax, cm, imgtype):
    if imgtype == '000':
        classes = ['sky', 'clouds', 'vegetation', 'built', 'windows', 'roads', 'cars', 'metal']
    else:
        classes = ['sky', 'clouds', 'vegetation', 'water', 'built', 'windows', 'roads', 'cars', 'metal']

    #fig, ax = plt.subplots()
    im = ax.imshow(cm, interpolation='nearest', cmap=plt.cm.Blues)
    ax.figure.colorbar(im, ax=ax)

    ax.set(xticks=np.arange(cm.shape[1]),
          yticks=np.arange(cm.shape[0]), 
          yticklabels=classes,
          ylabel='True Class',
          xlabel='Predicted Class',
          ylim=(cm.shape[0]-0.5, -0.5))
    ax.set_xticklabels(classes, rotation=45, ha='right', rotation_mode="anchor")

    thresh = cm.max()/2

    for i in range(cm.shape[0]):
        for j in range(cm.shape[1]):
            ax.text(j, i, format(cm[i,j], 'd'),
                   ha="center", va="center",
                   color="white" if cm[i,j] > thresh else "black")
    #fig.tight_layout()
    #plt.show()
```

## Model 1 vs Model 2 - train on each and test on each

```python
file_M1_1a = "./metrics/CNN_Model1_spatial_train_{0}_test_{0}_metrics.txt".format('108')
file_M1_1b = "./metrics/CNN_Model1_spatial_train_{0}_test_{0}_metrics.txt".format('000')
file_M1_2 = "./metrics/CNN_Model1_spatial_train_{0}_test_{0}_metrics.txt".format('north')

file_M2_1a = "./metrics/CNN_Model2_spatial_train_{0}_test_{0}_metrics.txt".format('108')
file_M2_1b = "./metrics/CNN_Model2_spatial_train_{0}_test_{0}_metrics.txt".format('000')
file_M2_2 = "./metrics/CNN_Model2_spatial_train_{0}_test_{0}_metrics.txt".format('north')

M1_1a = read_matrix(file_M1_1a, '108')
M1_1b = read_matrix(file_M1_1b, '000')
M1_2 = read_matrix(file_M1_2, 'north')
M2_1a = read_matrix(file_M2_1a, '108')
M2_1b = read_matrix(file_M2_1b, '000')
M2_2 = read_matrix(file_M2_2, 'north')
```

```python
plt.rcParams['font.family'] = 'DejaVu Sans'
plt.rcParams['font.size'] = 15

fig, ((ax1,ax2), (ax3,ax4), (ax5,ax6))= plt.subplots(3,2, figsize=(15,20))
plot_confusion_matrix(ax1, M1_1a, '108')
plot_confusion_matrix(ax2, M2_1a, '108')
plot_confusion_matrix(ax3, M1_1b, '000')
plot_confusion_matrix(ax4, M2_1b, '000')
plot_confusion_matrix(ax5, M1_2, 'north')
plot_confusion_matrix(ax6, M2_2, '108')

ax1.set_title('Model 1 - Scene 1-a - mF1=0.89')
ax2.set_title('Model 2 - Scene 1-a - mF1=0.97')
ax3.set_title('Model 1 - Scene 1-b - mF1=0.86')
ax4.set_title('Model 2 - Scene 1-b - mF1=0.92')
ax5.set_title('Model 1 - Scene 2 - mF1=0.82')
ax6.set_title('Model 2 - Scene 2 - mF1=0.90')

fig.tight_layout()
plt.show()
```

```python
fig.savefig("./plots/CNN_M1_M2_train_all_confusion_matrices.png", bbox_inches='tight', dpi=300)
```
