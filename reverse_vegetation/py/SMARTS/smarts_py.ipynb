## Reading SMARTS outputs

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
file = './SMARTS_295_Linux/smarts295.ext.txt'
extdf = pd.read_csv(file, sep=" ") 
```

```python
extdf.keys()
```

```python
fig = plt.subplots(figsize=(10,6))
for key in extdf.keys():
    if key != 'Wvlgth':
        plt.plot(extdf['Wvlgth'], extdf[key], label=key)

plt.legend()
plt.show()
```

```python
file = './SMARTS_295_Linux/smarts295.scn.txt'
scndf = pd.read_csv(file, sep=" ", header=2, skip_blank_lines=True, skipinitialspace=True) 
```

```python
scndf.keys()
```

```python
fig = plt.subplots(figsize=(10,6))
for key in scndf.keys():
    if key != 'WVLGTH':
        plt.plot(scndf['WVLGTH'], scndf[key], label=key)

plt.legend()
plt.show()
```
